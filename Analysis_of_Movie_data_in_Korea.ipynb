### Processing weekly boxoffice data

I retrieved weekly boxoffice data between week 1 of January 2009 and the last week of November 2018 and restored them in txt file.<br>
> it takes very long time to retrieve all data of last 10 year and Be aware that you can use an api key up to 3000 times. Also you can change the period

Here is the code processing __foreign commercial movies__ data

```python
from datetime import date,timedelta    #datetime package for calculating date
import json
import urllib.request


url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchWeeklyBoxOfficeList.json?key=################&multiMovieYn=N&repNationCd=F&weekGb=0&targetDt="
url_details = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key=################&movieCd="

# fill with your own api keys in ############## if you use this code.

current_date = date(2018,12,25)

file = open('FrnCmcialmovie.txt','w',encoding='UTF-8')
#Openning txt file to save data

movieCdlist = []
#List of movie codes

while True:
    str_date = date.isoformat(current_date).replace('-','')
    # replace '-' to blank
    
    url_open = urllib.request.urlopen(url+str_date)    #opening url added date
    json_read = url_open.read()
    data = json.loads(json_read)

    print('Retrieving',data['boxOfficeResult']['showRange'],'data...')    # notifying progress
    
    for movie_data in data['boxOfficeResult']['weeklyBoxOfficeList']:
        movieCd = movie_data['movieCd']
        audiAcc = movie_data['audiAcc']
        
        if movieCd in movieCdlist or int(movie_data['movieCd'][:4])<2009:
            continue
            
        movieCdlist.append(movieCd)
        
        url_details_open = urllib.request.urlopen(url_details+str(movieCd))    #opening url added date
        json_read = url_details_open.read()
        data = json.loads(json_read)
        
        movieNm = data['movieInfoResult']['movieInfo']['movieNm']
        movieNmEn = data['movieInfoResult']['movieInfo']['movieNmEn']
        
        genres = ''
        directors = ''
        actors = ''
        
        for genre in data['movieInfoResult']['movieInfo']['genres']:
            genres += genre['genreNm'] + ','
    
        for director in data['movieInfoResult']['movieInfo']['directors']:
            directors += director['peopleNm'] + ','
    
        for actor in data['movieInfoResult']['movieInfo']['actors']:
            actors += actor['peopleNm'] + ','
            if actors.count(',') == 9:
                break
        
        
        file.write('{}|{}|{}|{}|{}|{}|{}\n'.format(movieCd,movieNm,movieNmEn,audiAcc,genres[:-1],directors[:-1],actors[:-1])) 
        print(movieCd,movieNm,movieNmEn,audiAcc)
            
    print()
    
    
    current_date -= timedelta(days=7)  
    # minus a week
    
    if current_date < date(2009,1,1):    #if current date is before 2009, break it.
        break

file.close()

print('retreived',len(movieCdlist),'movies data')
```

### Audience count of commercial movies

Firstly, I defined a function which calculates counts and scores.<br> Commercial movies divided into __a million units__, and diversity movies divided into __one hundred thousand units__<br> Then, based on collected data, I visualized the difference of audiecne count.

```python
def scoring_cmcial(filename):  
    
    open_file = open(filename+'.txt','r',encoding='utf-8')
    read_file = open_file.readlines()
    audi_category = dict({'10M':0,'9M':0,'8M':0,'7M':0,'6M':0,'5M':0,'4M':0,'3M':0,'2M':0,'1M':0,'0M':0}) 
    audiAcc_array = np.array([])

    for line in read_file:

        movie_data = line.strip().split('|')
        code = movie_data[0]
        title = movie_data[1]
        title_Eng = movie_data[2]
        get_audiAcc = int(movie_data[3])
        genres = movie_data[4].split(',')
        directors = movie_data[5].split(',')
        actors = movie_data[6].split(',')

        audiAcc_array = np.insert(audiAcc_array, 0, get_audiAcc)
        movie_score = get_audiAcc//1000000                                  
        
        if movie_score > 10:
            movie_score = 10
        movie_category = str(movie_score) + 'M'
        audi_category[movie_category] = audi_category.get(movie_category, 0) + 1
        
        # Scores are given according to the number of audiences.
        for actor in actors:
            actors_score[actor] = actors_score.get(actor, 0) + movie_score
        for genre in genres:
            genres_score[genre] = genres_score.get(genre, 0) + movie_score
        for director in directors:
            directors_score[director] = directors_score.get(director, 0) + movie_score

    open_file.close()

    return audi_category, audiAcc_array

def scoring_dvsty(filename):
    
    open_file = open(filename+'.txt','r',encoding='utf-8')
    read_file = open_file.readlines()
    audi_category = dict({'1.0M':0,'0.9M':0,'0.8M':0,'0.7M':0,'0.6M':0,'0.5M':0,'0.4M':0,'0.3M':0,'0.2M':0,'0.1M':0,'0.0M':0}) 
    audiAcc_array = np.array([])

    for line in read_file:

        movie_data = line.strip().split('|')
        code = movie_data[0]
        title = movie_data[1]
        title_Eng = movie_data[2]
        get_audiAcc = int(movie_data[3])
        genres = movie_data[4].split(',')
        directors = movie_data[5].split(',')
        actors = movie_data[6].split(',')

        audiAcc_array = np.insert(audiAcc_array, 0, get_audiAcc)
        movie_score = get_audiAcc//100000                                    
        
        if movie_score > 10:
            movie_score = 10
        movie_category = str(movie_score/10) + 'M'
        audi_category[movie_category] = audi_category.get(movie_category, 0) + 1
        
        # Scores are given according to the number of audiences.
        for actor in actors:
            actors_dvsty_score[actor] = actors_dvsty_score.get(actor, 0) + movie_score
        for genre in genres:
            genres_dvsty_score[genre] = genres_dvsty_score.get(genre, 0) + movie_score
        for director in directors:
            directors_dvsty_score[director] = directors_dvsty_score.get(director, 0) + movie_score

    open_file.close()

    return audi_category, audiAcc_array
```
